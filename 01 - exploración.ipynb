```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')

# --- Cargar archivo ---
path = '/content/drive/MyDrive/DataProyectoIA/train.csv'
df = pd.read_csv(path, sep=",")

# Vista rápida de los datos
print("Dimensiones del dataset:", df.shape)
print(df.head())
print("\nColumnas disponibles:", df.columns.tolist())

# --- Conteo de estudiantes por programa académico ---
conteo_programa = df['E_PRGM_ACADEMICO'].value_counts()
print("\nEstudiantes por programa académico:\n", conteo_programa)

# --- Conteo por departamento ---
conteo_depto = df['E_PRGM_DEPARTAMENTO'].value_counts()
print("\nEstudiantes por departamento:\n", conteo_depto)

plt.figure(figsize=(10,5))
sns.countplot(
    y='E_PRGM_DEPARTAMENTO',
    data=df,
    order=df['E_PRGM_DEPARTAMENTO'].value_counts().index,
    color='steelblue'
)
plt.title("Número de estudiantes por departamento")
plt.show()

# --- Distribución del rendimiento global ---
plt.figure(figsize=(6,4))
sns.histplot(df['RENDIMIENTO_GLOBAL'], bins=20, kde=True, color='blue')
plt.title("Distribución del Rendimiento Global")
plt.show()

# --- Relación entre Estrato de Vivienda y Rendimiento ---
plt.figure(figsize=(8,5))
sns.boxplot(
    x='F_ESTRATOVIVIENDA',
    y='RENDIMIENTO_GLOBAL',
    data=df,
    color='lightgreen'
)
plt.title("Rendimiento Global vs Estrato de Vivienda")
plt.show()
```
